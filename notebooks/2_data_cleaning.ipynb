```python
import pandas as pd
import numpy as np
```

```python
orders = pd.read_csv("../data/raw/olist_orders_dataset.csv")
customers = pd.read_csv("../data/raw/olist_customers_dataset.csv")
payments = pd.read_csv("../data/raw/olist_order_payments_dataset.csv")
items = pd.read_csv("../data/raw/olist_order_items_dataset.csv")
```

```python
date_cols = [
    'order_purchase_timestamp',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
```

```python
for col in date_cols:                     #Converting columns to datetime
    orders[col] = pd.to_datetime(orders[col])
```

```python
orders.isna().sum() # missing
```

```python
orders = orders.dropna(subset = ["order_delivered_customer_date"])
```

```python
orders.isna().sum()
```

```python
#merging data sets
orders_customers = orders.merge(customers, on = "customer_id", how = "left")
```

```python
orders_items = orders_customers.merge(items, on ="order_id",how = "left")
```

```python
orders_final = orders_items.merge(payments, on = "order_id", how="left")
```

```python
orders_final.head()
```

```python
orders_final['total_order_value'] = (
    orders_final['price'] + orders_final['freight_value']
)
```

```python
orders_items.head()
```

```python
orders_final.to_csv("../data/cleaned/cleaned_orders.csv", index=False)
```
