## Purpose of This Notebook

This notebook summarizes the key insights derived from the exploratory data analysis
and presents them in a clear, business-friendly manner.

The objective is to:
- Highlight meaningful patterns in customer spending and payment behavior
- Translate data findings into actionable business insights
- Support decision-making without requiring technical background


## Data Context

The analysis is based on the Brazilian E-Commerce Public Dataset by Olist.

Key entities used:
- Orders
- Order items (price and freight)
- Payments
- Customers

The primary metric analyzed is **Total Order Value**, calculated as:
> Item price + freight value


## Insight 1: Most Orders Are Low-Value with a Long Tail of High-Value Purchases


- The majority of customer orders fall into a low-to-mid price range
- A small proportion of orders account for very high purchase values




## Insight 2: Payment Method Is Strongly Associated with Order Value


- Credit card payments are associated with higher median and average order values
- Voucher and debit card payments tend to be used for lower-value purchases




## Insight 3: Installments Enable Higher-Value Purchases


- Average order value increases as the number of payment installments increases
- Customers purchasing expensive items prefer spreading payments over time



## Insight 4: Revenue Is Concentrated in a Few States

**What we observed**
- A small number of states contribute a large share of total revenue
- Demand is not uniformly distributed across regions





## Predictive Insight: Identifying High-Value Orders Early

A simple classification model was built to predict whether an order is likely to be
high-value based on payment behavior.

**Key takeaway**
- Payment installments and payment type are strong indicators of high-value orders
